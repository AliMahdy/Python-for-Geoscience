```python
# --------------------------------------------
# 8. CONDITIONAL OPERATIONS
# --------------------------------------------
print("\n8. CONDITIONAL OPERATIONS")
print("-" * 40)

# np.where - classify reservoir quality
reservoir_quality = np.where(
    (porosity > 15) & (permeability > 50),
    'Good', 'Poor'
)
print("Reservoir quality classification:")
for i in range(len(formations)):
    print(f"  {formations[i]}: {reservoir_quality[i]}")
print()

# np.select - multiple conditions
rock_quality = np.select(
    [
        (porosity >= 20) & (permeability >= 100),
        (porosity >= 15) & (permeability >= 50),
        (porosity < 15) | (permeability < 10)
    ],
    ['Excellent', 'Good', 'Poor'],
    default='Fair'
)
print("Detailed rock quality classification:")
for i in range(len(formations)):
    print(f"  {formations[i]}: {rock_quality[i]} (φ={porosity[i]}%, k={permeability[i]} mD)")
print()
```

```python
# --------------------------------------------
# 9. HANDLING MISSING DATA
# --------------------------------------------
print("\n9. HANDLING MISSING DATA")
print("-" * 40)

# Create array with NaN (Not a Number) for missing log values
porosity_with_nan = np.array([22.5, np.nan, 18.7, 25.1, 6.2])
print("Porosity with missing value (NaN):")
print(porosity_with_nan)
print()

# Check for NaN
print("Which values are NaN?")
print(np.isnan(porosity_with_nan))
print()

# Count NaN values
print(f"Number of missing values: {np.sum(np.isnan(porosity_with_nan))}")
print()

# Calculate statistics ignoring NaN
print(f"Mean porosity (ignoring NaN): {np.nanmean(porosity_with_nan):.2f} %")
print(f"Std dev (ignoring NaN): {np.nanstd(porosity_with_nan):.2f} %")
print()

# Fill NaN with mean value
filled_porosity = np.where(
    np.isnan(porosity_with_nan),
    np.nanmean(porosity_with_nan),
    porosity_with_nan
)
print("Porosity with NaN filled by mean:")
print(filled_porosity)
print()
```

```python
# --------------------------------------------
# 10. PRACTICAL PETROPHYSICAL CALCULATIONS
# --------------------------------------------
print("\n10. PRACTICAL PETROPHYSICAL CALCULATIONS")
print("-" * 40)

# Calculate Storage Capacity (φ × k)
storage_capacity = porosity * permeability
print("Storage Capacity (φ × k) for each formation:")
for i in range(len(formations)):
    print(f"  {formations[i]}: {storage_capacity[i]:.2f}")
print()

# Calculate Flow Potential (simplified)
# Higher porosity AND permeability = better flow
flow_potential = (porosity / 30) * np.log10(permeability + 0.001)
print("Normalized Flow Potential Index:")
for i in range(len(formations)):
    print(f"  {formations[i]}: {flow_potential[i]:.3f}")
print()

# Identify pay zones (threshold-based)
is_pay_zone = (porosity > 15) & (permeability > 30)
pay_formations = formations[is_pay_zone]
print("Identified Pay Zones (φ > 15% AND k > 30 mD):")
print(pay_formations)
print()

# Calculate average properties for pay zones only
if np.any(is_pay_zone):
    print(f"Average Porosity in Pay Zones: {np.mean(porosity[is_pay_zone]):.2f} %")
    print(f"Average Permeability in Pay Zones: {np.mean(permeability[is_pay_zone]):.2f} mD")
print()
```
