## 🧮 Part 2: Python as Your Geological Calculator


# Basic arithmetic: Calculate fold wavelength from outcrop measurements
# Formula: λ = 2π × (amplitude / strain_rate)^(1/2)

```python
import math  # Built-in math library - always available!

amplitude_m = 15.5      # Fold amplitude in meters
strain_rate = 0.08      # Dimensionless strain rate

wavelength_m = 2 * math.pi * math.sqrt(amplitude_m / strain_rate)
print(f"Fold wavelength: {wavelength_m:.2f} meters")
```


# Geological constants in action
# Example: Calculate hydrostatic pressure at depth
# Formula: P = ρ × g × h


```python
rho_water = 1000        # kg/m³ (density of water)
g = 9.81                # m/s² (gravity)
depth_m = 1250          # meters below surface

pressure_pa = rho_water * g * depth_m
pressure_mpa = pressure_pa / 1e6  # Convert to MPa

print(f"Hydrostatic pressure at {depth_m} m depth:")
print(f"  → {pressure_pa:,.0f} Pascals")
print(f"  → {pressure_mpa:.2f} MPa")
print(f"\n💡 Rule of thumb: ~0.01 MPa per meter of water depth")
```


# Real challenge: Calculate volume of a cylindrical salt dome
# Formula: V = π × r² × h

```python
radius_km = 3.2         # Radius of salt dome (km)
height_km = 1.8         # Vertical extent (km)

volume_km3 = math.pi * (radius_km ** 2) * height_km
volume_m3 = volume_km3 * 1e9  # Convert km³ → m³

print("SALT DOME VOLUME CALCULATION")
print("=" * 40)
print(f"Radius: {radius_km} km")
print(f"Height: {height_km} km")
print(f"Volume: {volume_km3:.2f} km³")
print(f"       = {volume_m3:.2e} m³")
print(f"\n💡 Context: This dome could store ~{volume_km3*0.2:.1f} km³ of CO₂")
print("   (assuming 20% pore space for carbon sequestration)")
```


 ## ⛏️ Part 5: Real Challenge – Structural Geology Application

# PROBLEM: Calculate volume and surface area of a conical volcanic edifice
#
# Given field measurements:
#   - Base diameter: 8.4 km
#   - Height: 2.1 km
#   - Erosion has removed top 15% of original height

```python

import math

# Field measurements
base_diameter_km = 8.4
current_height_km = 2.1

# Calculate original height before erosion
erosion_fraction = 0.15
original_height_km = current_height_km / (1 - erosion_fraction)

# Calculate radius
radius_km = base_diameter_km / 2

# Volume of cone: V = (1/3) × π × r² × h
current_volume_km3 = (1/3) * math.pi * (radius_km ** 2) * current_height_km
original_volume_km3 = (1/3) * math.pi * (radius_km ** 2) * original_height_km
eroded_volume_km3 = original_volume_km3 - current_volume_km3
# Surface area of cone (lateral): A = π × r × √(r² + h²)
slant_height_km = math.sqrt(radius_km**2 + current_height_km**2)
surface_area_km2 = math.pi * radius_km * slant_height_km

print("🌋 VOLCANIC EDIFICE ANALYSIS")
print("=" * 60)
print(f"Base diameter:       {base_diameter_km} km")
print(f"Current height:      {current_height_km} km")
print(f"Original height:     {original_height_km:.2f} km (before 15% erosion)")
print("-" * 60)
print(f"CURRENT VOLUME:      {current_volume_km3:.2f} km³")
print(f"ERODED VOLUME:       {eroded_volume_km3:.2f} km³")
print(f"TOTAL ORIGINAL VOL:  {original_volume_km3:.2f} km³")
print(f"SURFACE AREA:        {surface_area_km2:.2f} km²")
print("=" * 60)
print(f"\n💡 Interpretation: {eroded_volume_km3:.1f} km³ of material")
print("   has been removed by erosion—equivalent to ~200 years of")
print("   sediment deposition in a major delta system!")
```

## 🧪 Part 6: Hands-On Exercise – Core Plug Analysis

# SYNTHETIC CORE PLUG DATASET (no external files needed!)
# Measurements from 8 core plugs extracted at different depths

```python
depth_m = [1200.5, 1201.0, 1201.5, 1202.0, 1202.5, 1203.0, 1203.5, 1204.0]
bulk_density_gcm3 = [2.45, 2.51, 2.48, 2.62, 2.59, 2.47, 2.53, 2.50]
porosity_percent = [18.2, 15.5, 16.8, 8.5, 9.7, 17.9, 14.3, 16.1]
permeability_md = [125, 98, 112, 18, 24, 135, 87, 105]

print("CORE PLUG ANALYSIS - SYNTHETIC DATASET")
print("=" * 70)
print(f"{'Depth (m)':<12} {'Bulk Density':<15} {'Porosity':<12} {'Permeability':<15}")
print(f"{'':<12} {'(g/cm³)':<15} {'(%)':<12} {'(mD)':<15}")
print("-" * 70)

for i in range(len(depth_m)):
    print(f"{depth_m[i]:<12.1f} {bulk_density_gcm3[i]:<15.2f} {porosity_percent[i]:<12.1f} {permeability_md[i]:<15.0f}")
```

```python
# EXERCISE 1: Calculate average reservoir properties
avg_porosity = sum(porosity_percent) / len(porosity_percent)
avg_permeability = sum(permeability_md) / len(permeability_md)

print("📊 RESERVOIR PROPERTY STATISTICS")
print(f"   Average porosity:    {avg_porosity:.1f}%")
print(f"   Average permeability: {avg_permeability:.0f} mD")
```

```python
# EXERCISE 2: Identify best reservoir intervals (porosity > 15% AND permeability > 100 mD)
pay_zones = []
for i in range(len(depth_m)):
    if porosity_percent[i] > 15.0 and permeability_md[i] > 100:
        pay_zones.append({
            'depth': depth_m[i],
            'porosity': porosity_percent[i],
            'perm': permeability_md[i]
        })

print(f"\n💎 PAY ZONES IDENTIFIED: {len(pay_zones)} intervals")
print("-" * 50)
for zone in pay_zones:
    print(f"   Depth {zone['depth']:.1f} m → φ={zone['porosity']:.1f}%, k={zone['perm']} mD")
```

```python
# EXERCISE 3: Unit conversion challenge (industry requirement!)
# Convert all depths from meters to feet (1 m = 3.28084 ft)

depth_ft = [d * 3.28084 for d in depth_m]

print("\n📐 UNIT CONVERSION: Meters → Feet")
print(f"{'Depth (m)':<12} {'Depth (ft)':<12}")
print("-" * 25)
for i in range(len(depth_m)):
    print(f"{depth_m[i]:<12.1f} {depth_ft[i]:<12.1f}")

# Bonus: Calculate thickness of pay zone in feet
if len(pay_zones) >= 2:
    top_pay = pay_zones[0]['depth']
    base_pay = pay_zones[-1]['depth']
    net_pay_m = base_pay - top_pay
    net_pay_ft = net_pay_m * 3.28084
    print(f"\n✅ NET PAY THICKNESS: {net_pay_m:.1f} m = {net_pay_ft:.1f} ft")
```
